# Explore MODIS data 

We have some `.tif` files provided by ECMWF

This is MODIS data obtained from [Google Earth Engine](https://developers.google.com/earth-engine/datasets/catalog/MODIS_006_MYD11A1), having undergone some processing. 

It is the `MYD11A1.006` product. Note that `MYD` refers to Aqua/MODIS (c.f. Terra). 

The corresponding dataset can also be found in the [datapool](https://lpdaac.usgs.gov/products/myd11a1v006/) - the link from Google Earth seems to take you to Terra?

It is a level 3 product, derived from the [level 2 swath product](https://lpdaac.usgs.gov/products/myd11_l2v006/).

It is also a version 6.0 product - [newer 6.1 products do exist](https://lpdaac.usgs.gov/news/upcoming-release-lp-daac-modis-version-61-data-products/)

See [here](https://modaps.modaps.eosdis.nasa.gov/services/about/nomenclature.html) for some nomenclature, and also check out the [Users's guide](https://lpdaac.usgs.gov/documents/118/MOD11_User_Guide_V6.pdf)

---

Lets have a look at the data provided by Mat, which is Terra/MODIS for a particular day and the corresponding `Day_view_time` which is the time of the observation in **local solar time**

```python
import xarray as xr
f = '/network/group/aopp/predict/TIP016_PAXTON_RPSPEEDY/ML4L/ECMWF_files/terraDayTime_errorGTE03K_04km_2018-01-01.tif'
t = xr.open_rasterio(f)
display(t)
```

Lets turn it into a pandas df and plot it:

```python
t_df = t.to_dataframe(name='local_solar_time_uncorrected')
t_df['local_solar_time'] = t_df['local_solar_time_uncorrected']/0.02*0.1 #by hand correction
t_df = t_df.dropna() #Get rid of nulls

```

```python
display (t_df)
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import pandas as pd
def plotit(df,x):
    
    #Pass a df object, convert to xarray object f and the value to plot x
    
    
    
    f = xr.Dataset.from_dataframe(df)
    f1 = df.to_xarray()
    
    display(f)
    display(f1)
    
    
    
    
    
    fig = plt.figure(figsize=[24,12])
    ax = fig.add_subplot(111, projection=ccrs.PlateCarree(central_longitude=0))

    f1[x].plot(ax=ax, 
             cmap='jet',
             transform=ccrs.PlateCarree())
    ax.coastlines()


    plt.show()
```

```python
plotit(t_df,'local_solar_time')
```

Rather than local solar time, we want to use some 'absolute time' which we can then link to ERA5 data. Since ERA5 has a precision of hours, we just need this absolute time to a precision of hours too.

The conversion between local solar time (LST) and UTC [from the docs](https://lpdaac.usgs.gov/documents/118/MOD11_User_Guide_V6.pdf) is 

$$ \text{LST} = \text{UTC} + \frac{\text{longitude [deg]}}{15}$$


```python
t_df['UTC'] = t_df['local_solar_time'] - t_df.index.get_level_values('x')/15.0
```

```python
plotit(t_df,'UTC')
```

```python
display(t_df['UTC'].idxmax(),t_df['UTC'].max())
display(t_df['UTC'].idxmin(),t_df['UTC'].min())

```

```python
display(t_df['local_solar_time'].idxmax(),t_df['local_solar_time'].max())
display(t_df['local_solar_time'].idxmin(),t_df['local_solar_time'].min())
```

```python
t_df_exclude_poles = t_df.query('-60.0 < y < 60.0')
display(t_df_exclude_poles['UTC'].idxmax(),t_df_exclude_poles['UTC'].max())
display(t_df_exclude_poles['UTC'].idxmin(),t_df_exclude_poles['UTC'].min())
```
