# Check time join for ERA/MODIS


Our current appraoach for joining the ERA and MODIS data in time is to first select an hourly slice of MODIS data.

We do this as follows:

* Load a 24 hour file of MODIS data

* Map longitude to UTC, assuming a fixed local solar time of 13.30, rounded to nearest hour

* Select only times you are interested in e.g. all data with UTC = 11


There are also some additional considerations w.r.t the dateline if the time you are loading is outside of the min/max hours. For now we can ignore this subtlety 

---


The fundamental assumption here is that we can map longtiude to time in this way. This assumption is true if all observations are taken at a fixed local solar time of 13.30 (for AquaDay). **Is this true?**


## 1. Load some data

Lets get a day of MODIS data. We will use the [MYD11C1](https://lpdaac.usgs.gov/products/myd11c1v061/) data product. See also [the docs](https://lpdaac.usgs.gov/documents/118/MOD11_User_Guide_V6.pdf).

We can then load this data and take a quick look:

```python
import xarray as xr
import rioxarray as rxr

modis_xarray= rxr.open_rasterio('/network/group/aopp/predict/TIP016_PAXTON_RPSPEEDY/ML4L/example_MODIS_data.hdf',masked=True)
display(modis_xarray)
```

## 2. Tidy up data

We are not interested in all these variables. Lets grab just the day view temperature and time, along with the lat/lon position

```python
modis_df = modis_xarray[['LST_Day_CMG','Day_view_time']].to_dataframe().reset_index().dropna().drop(['band','spatial_ref'],axis=1)
```

```python
modis_df
```

We can also convert the tempearture and time to some nicer units:

```python
modis_df['temperature_kelvin'] = modis_df['LST_Day_CMG']*modis_xarray.LST_Day_CMG.scale_factor
modis_df['local_solar_time'] = modis_df['Day_view_time']*modis_xarray.Day_view_time.scale_factor
```

```python
modis_df
```

In practice we know there are issues at the "edges" i.e. extrema of longitude and also at the poles. Lets restrict our analysis to a subset of the globe:

```python
modis_df_filtered = modis_df.query('-60 < y < 60 & -150 < x < +150').copy()
```

```python
modis_df_filtered
```

We can see just from reading the column that it is not the case that the local solar time always equals 13.30 for this dataset.

What does this imply when we go to calculate UTC?

```python
modis_df_filtered['UTC_theoretical'] = 13.5 - modis_df_filtered.x/15                           # this is how we currently define the absolute time
modis_df_filtered['UTC_actual'] = modis_df_filtered.local_solar_time - modis_df_filtered.x/15  # this is the **actual** absolute time at which the observation was taken

modis_df_filtered['UTC_error'] = modis_df_filtered.UTC_actual - modis_df_filtered.UTC_theoretical
```

## 3. Make some plots

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mc
import matplotlib.colorbar as cb 
def plot_map(df,z,vmin,vmax,cmap):
    
    
    """
    Hacky plotting function that just plots as scatter
    """
    
    x = df.x.values
    y = df.y.values
    z1 = df[z].values

    
    #Scatter plot it
    fig,[ax,cax] = plt.subplots(1,2, gridspec_kw={"width_ratios":[50,1]},figsize=(30, 20))

    
    norm = mc.Normalize(vmin=vmin, vmax=vmax)

    cb1 = cb.ColorbarBase(cax, cmap=cmap,norm=norm,orientation='vertical')
    
    
    sc = ax.scatter(x, y,s=1,c=cmap(norm(z1)),linewidths=1, alpha=1)

    plt.show()
    
    
    
```

```python
modis_df_filtered
```

```python
plot_map(modis_df_filtered,'UTC_actual',vmin=0,vmax=24,cmap=plt.cm.coolwarm)
```

```python
plot_map(modis_df_filtered,'UTC_theoretical',vmin=0,vmax=24,cmap=plt.cm.coolwarm)
```

```python
plot_map(modis_df_filtered,'UTC_error',vmin=modis_df_filtered.UTC_error.min(),vmax=modis_df_filtered.UTC_error.max(),cmap=plt.cm.bwr)
```
