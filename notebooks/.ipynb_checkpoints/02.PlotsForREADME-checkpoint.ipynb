# Example plot showing a strip of joined MODIS/ERA data

```python
import pandas as pd
#Load a month of joined data
f1 = '/network/group/aopp/predict/TIP016_PAXTON_RPSPEEDY/ML4L/ECMWF_files/raw/processed_data/joined_data/MODIS_ERA_2016_01.parquet'
example_month = pd.read_parquet(f1,columns=['latitude_ERA','longitude_ERA', 'MODIS_LST','time'])

#Just get an hour
times = example_month.time.unique()
t = times[8]
df_hour = example_month.query(f'time == "{t}"')
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as crs
import cartopy.feature as cfeature
import matplotlib.colors as mc


fig = plt.figure(figsize=(20,16))

ax = fig.add_subplot(1,1,1, projection=crs.Robinson())

ax.set_global()

border_colors = 'black'
ax.add_feature(cfeature.COASTLINE, edgecolor=border_colors)
ax.add_feature(cfeature.BORDERS, edgecolor=border_colors)
ax.gridlines()
ax.stock_img()

vmin=df_hour.MODIS_LST.min()
vmax=df_hour.MODIS_LST.max()
norm = mc.Normalize(vmin=vmin, vmax=vmax)

cmap=plt.cm.magma
sc = plt.scatter(x=df_hour.longitude_ERA, y=df_hour.latitude_ERA,
            color = cmap(norm(df_hour.MODIS_LST)),
            #color="dodgerblue",
            s=1,
            alpha=0.5,
            cmap=cmap,
            transform=crs.PlateCarree()) ## Important

#Colorbar settings
plt.clim(vmin, vmax) #cb not picking up these limits from `sc` automatically?
plt.colorbar(sc,orientation='horizontal',fraction=0.026, pad=0.01,cmap=cmap)
plt.set_cmap(cmap)
plt.savefig('media/example_joining_strip.png',bbox_inches='tight', pad_inches=0, dpi=300)
plt.show()
```
