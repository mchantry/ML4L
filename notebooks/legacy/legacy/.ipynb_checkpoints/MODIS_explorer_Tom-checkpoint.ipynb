```python
import xarray as xr
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
time_terra_grid = ds = xr.open_rasterio('terraDayTime_errorGTE03K_04km_2018-01-01.tif')
```

```python
time_terra_grid_2 = ds = xr.open_rasterio('terraDayTime_errorGTE03K_04km_2018-01-02.tif')
```

```python
plt.figure(dpi=200)
plt.contourf(time_terra_grid[0,:,:]/0.02*0.1,levels=np.arange(9.5,12.6,0.1))
plt.colorbar()
```

```python
plt.figure(dpi=200)
plt.contourf(time_terra_grid_2[0,:,:]/0.02*0.1,levels=np.arange(9.5,12.6,0.1))
plt.colorbar()
```

```python
plt.figure(dpi=200)
plt.contourf((time_terra_grid[0,:,:]-time_terra_grid_2[0,:,:])/0.02*0.1,levels=np.arange(-1.5,1.6,0.1),
            cmap='RdBu')
plt.colorbar()
```
