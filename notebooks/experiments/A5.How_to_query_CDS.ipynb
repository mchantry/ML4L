# How to query CDS to get data on the correct grid


A trick from Mat Chantry

```python
import cdsapi
```

```python

savedir = '/network/group/aopp/predict/TIP016_PAXTON_RPSPEEDY/ML4L/ECMWF_files/raw/NewEraFields/'
c = cdsapi.Client()
times = [f'{i:02d}:00:00' for i in range(24)]
c.retrieve("reanalysis-era5-complete",
           {  
               'type'    : 'an',
               'param'   : '235',
               'levtype' : 'sfc',
               'date'    : '2016-01-01/to/2016-01-31',
               'time'    :  times,
               'format'  : 'grib', # Default: grib
           }, 
           savedir+'output.grib')
```

```python
import xarray as xr
```

```python

ds = xr.open_dataset(savedir+'output.grib',engine='cfgrib',backend_kwargs={'indexpath': ''})

```
